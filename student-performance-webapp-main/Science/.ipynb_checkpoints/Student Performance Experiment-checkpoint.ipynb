# Student Performance Analysis and Prediction

## Introduction
This project aims to analyze and predict student performance based on various factors, including demographic details, study habits, parental involvement, and participation in extracurricular activities. By understanding these influences, we can identify patterns and provide actionable insights to improve educational outcomes.


## Table of Contents

* Student Information
* Demographic Details
* Study Habits
* Parental Involvement
* Extracurricular Activities
* Academic Performance
* Target Variable: Grade Class

1. ### Student Information
* StudentID: A unique identifier assigned to each student (1001 to 3392).

2. ### Demographic Details
* Age: The age of the students ranges from 15 to 18 years.
* Gender: Gender of the students, where 0 represents Male and 1 represents Female.
* Ethnicity: The ethnicity of the students, coded as follows:
>0: Caucasian
1: African American
2: Asian
3: Other
* ParentalEducation: The education level of the parents, coded as follows:
> 0: None
1: High School
2: Some College
3: Bachelor's
4: Higher

3. ### Study Habits
* StudyTimeWeekly: Weekly study time in hours, ranging from 0 to 20.
* Absences: Number of absences during the school year, ranging from 0 to 30.
* Tutoring: Tutoring status, where 0 indicates No and 1 indicates Yes.

4. ### Parental Involvement
* ParentalSupport: The level of parental support, coded as follows:
> 0: None
1: Low
2: Moderate
3: High
4: Very High

5. ### Extracurricular Activities
* Extracurricular: Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.
* Sports: Participation in sports, where 0 indicates No and 1 indicates Yes.
* Music: Participation in music activities, where 0 indicates No and 1 indicates Yes.
* Volunteering: Participation in volunteering, where 0 indicates No and 1 indicates Yes.

6. ### Academic Performance
* GPA: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities.
Target Variable: Grade Class
* GradeClass: Classification of students' grades based on GPA:
>0: 'A' (GPA >= 3.5)
1: 'B' (3.0 <= GPA < 3.5)
2: 'C' (2.5 <= GPA < 3.0)
3: 'D' (2.0 <= GPA < 2.5)
4: 'F' (GPA < 2.0)

## Libraries Importation
We begin by importing essential libraries for data manipulation, visualization, and modeling. These include pandas, numpy, matplotlib, seaborn, and several Scikit-learn modules.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style='whitegrid')
```

## Data Collection and Loading
The dataset is loaded into a pandas DataFrame. This dataset contains various attributes related to students, such as demographic details, study habits, parental involvement, extracurricular activities, and academic performance.

```python
# Load the dataset
file_path = 'Student_performance_data.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
display(data.head())
display(data.tail())
```

## Data Cleaning
Data cleaning involves handling missing values and ensuring that the data types of the features are appropriate. This step is crucial for ensuring the accuracy and reliability of our analysis and modeling.

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Convert categorical columns to appropriate data types
data['Gender'] = data['Gender'].astype('category')
data['Ethnicity'] = data['Ethnicity'].astype('category')
data['ParentalEducation'] = data['ParentalEducation'].astype('category')
data['Tutoring'] = data['Tutoring'].astype('category')
data['ParentalSupport'] = data['ParentalSupport'].astype('category')
data['Extracurricular'] = data['Extracurricular'].astype('category')
data['Sports'] = data['Sports'].astype('category')
data['Music'] = data['Music'].astype('category')
data['Volunteering'] = data['Volunteering'].astype('category')
data['GradeClass'] = data['GradeClass'].astype('category')
```

## Exploratory Data Analysis (EDA)
Exploratory Data Analysis helps us understand the distribution and relationships within the data. We use visualizations like correlation heatmaps and distribution plots to identify patterns and insights that will guide our modeling approach.


```python
# Summary statistics
data.describe()
```

### What is the distribution of GPA among students?

```python
# Plot the distribution of GPA
plt.figure(figsize=(10, 6))
sns.histplot(data['GPA'], kde=True, bins=30, color='blue')
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for the 'GradeClass' column
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='GradeClass', data=data, palette='viridis')

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

for bar, color in zip(ax.patches, colors):
    bar.set_color(color)

# Custom Legend
legend_labels = ['Grade A', 'Grade B', 'Grade C', 'Grade D', 'Grade F']
custom_legend = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]
plt.legend(custom_legend, legend_labels, title='Grade Classes')

# Adding titles and labels
plt.title('Grade Class Distribution of Students')
plt.xlabel('Grade Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

### What is the average GPA and grade for each gender?

```python
average_gpa_gender = data.groupby('Gender')['GPA'].mean()
average_gpa_gender
```

```python
# Plotting the average GPA by Parental Education Level

plt.figure(figsize=(12, 6))
average_gpa_gender.plot(kind='bar')
plt.title('Average GPA by Gender')
plt.xlabel('Gender')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Mapping of codes to labels
gender_labels = {
    0: 'Male',
    1: 'Female',
}
# Plot the relationship between GradeClass and Gender with custom legend
plt.figure(figsize=(12, 6))
sns.countplot(x='GradeClass', hue='Gender', data=data, palette='viridis')
plt.title('Relationship between Grade Class and Gender')
plt.xlabel('Grade Class')
plt.ylabel('Count')

# Create custom legend
legend_labels = [f'{key}: {value}' for key, value in gender_labels.items()]
plt.legend(title='Parental Support', labels=legend_labels)

plt.show()
```

### How does parental education level and Parental Support affect student GPA?

```python
average_gpa_parentaled = data.groupby('ParentalEducation')['GPA'].mean()
average_gpa_parentaled
```

```python
# Plotting the average GPA by Parental Education Level

plt.figure(figsize=(12, 6))
average_gpa_parentaled.plot(kind='bar')
plt.title('Average GPA by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Mapping of codes to labels
parental_support_labels = {
    0: 'None',
    1: 'Low',
    2: 'Moderate',
    3: 'High',
    4: 'Very High'
}

# Plot the relationship between GradeClass and ParentalSupport with custom legend
plt.figure(figsize=(12, 6))
sns.countplot(x='GradeClass', hue='ParentalSupport', data=data, palette='viridis')
plt.title('Relationship between Grade Class and Parental Support')
plt.xlabel('Grade Class')
plt.ylabel('Count')

# Create custom legend
legend_labels = [f'{key}: {value}' for key, value in parental_support_labels.items()]
plt.legend(title='Parental Support', labels=legend_labels)

plt.show()
```

### Do students who participate in extracurricular activities have higher GPAs?

```python
average_gpa_extracurricular = data.groupby('Extracurricular')['GPA'].mean()
average_gpa_extracurricular
```

```python
plt.figure(figsize=(12, 6))
average_gpa_extracurricular.plot(kind='bar')
plt.title('Average GPA by Extracurricular Activity')
plt.xlabel('Extracurricular Activity')
plt.ylabel('Average GPA')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Mapping of codes to labels
extracurricular_labels = {
    0: 'No',
    1: 'Yes',
}

# Plot the relationship between GradeClass and ParentalSupport with custom legend
plt.figure(figsize=(12, 6))
sns.countplot(x='GradeClass', hue='Extracurricular', data=data, palette='viridis')
plt.title('Relationship between Grade Class and Extracurricular Activity')
plt.xlabel('Grade Class')
plt.ylabel('Count')

# Create custom legend
legend_labels = [f'{key}: {value}' for key, value in extracurricular_labels.items()]
plt.legend(title='Parental Support', labels=legend_labels)

plt.show()
```

### What is the distribution of study time and absences among students?

```python
# Plot the distribution of study time
plt.figure(figsize=(10, 6))
sns.histplot(data['StudyTimeWeekly'], kde=True, bins=30, color='blue')
plt.title('Distribution of Study Time')
plt.xlabel('Study Time')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of A
plt.figure(figsize=(10, 6))
sns.histplot(data['Absences'], kde=True, bins=30, color='blue')
plt.title('Distribution of Absences')
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.show()
```

### Is there a relationship between study time, Absences, Tutorinng and GPA?

```python
# Plot Study Time vs GPA
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='StudyTimeWeekly', y='GPA', line_kws={'color': 'red'})
plt.title('Study Time vs GPA')
plt.xlabel('Study Time Weekly (hours)')
plt.ylabel('GPA')
plt.show()
```

```python
correlation_study_time_gpa = data['StudyTimeWeekly'].corr(data['GPA'])
display(correlation_study_time_gpa)
```

```python
# Plot Absences vs GPA
plt.figure(figsize=(10, 6))
sns.regplot(x='Absences', y='GPA', data=data, line_kws={'color': 'red'})
plt.title('Absences vs GPA')
plt.xlabel('Number of Absences')
plt.ylabel('GPA')
plt.show()
```

```python
correlation_absences_gpa = data['Absences'].corr(data['GPA'])
display(correlation_absences_gpa)
```

```python
# Mapping of codes to labels
tutoring_labels = {
    0: 'No',
    1: 'Yes',
}

# Plot the relationship between GradeClass and Tutoring with custom legend
plt.figure(figsize=(12, 6))
sns.countplot(x='GradeClass', hue='Tutoring', data=data, palette='viridis')
plt.title('Relationship between Grade Class and Tutoring')
plt.xlabel('Grade Class')
plt.ylabel('Count')

# Create custom legend
legend_labels = [f'{key}: {value}' for key, value in tutoring_labels.items()]
plt.legend(title='Parental Support', labels=legend_labels)

plt.show()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(14, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## Analysis Summary

The analysis of the student performance dataset reveals significant influences on GPA:

- **Increased Study Time**: Students who spend more time studying each week tend to have higher GPAs.
- **Participation in Extracurricular Activities**: Engagement in extracurricular activities such as sports, music, and volunteering is positively correlated with GPA.
- **Attendance**: Higher numbers of absences are associated with lower GPAs.

These insights underscore the importance of consistent attendance, and balanced extracurricular engagement in fostering academic success.
